```python
# Import DB instance and connect to the Database
from db import DB, DemoDB

# We are creating a DB in memory.  It doesn't exist as a file.
# To create the SQLite database as a file, just replace with a real filename
db = DB(filename=':memory:', dbtype="sqlite")
```

```python
# Let's Readin Titanic Dataset 
titanic = pd.read_csv('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv')
```

```python
titanic.head(5)
```

```python
# Take a look at the columns
titanic.info()
```

```python
# Write to Database.  We will then use to run queries
con = db.con
titanic.to_sql('titanic', con)
```

```python
# Select 5 rows
# Semicolan is optional, but required if you query directly in sqlite. So, I add it as good practice here

db.query('SELECT * FROM titanic LIMIT 5;')
```

```python
# Select All people who were over 70 years old
db.query('''SELECT * FROM titanic WHERE age > 70''')
```

```python
# Total Count of how many Survived
db.query('SELECT count(*) as survival_count from titanic where survived = 1')
```

```python
# How many did not survive
db.query('SELECT count(*) as survival_count from titanic where survived = 0')
```

```python
# How many survived who were First Class
db.query('SELECT count(*) as survival_count from titanic where survived = 1 and pclass = 1')
```

```python
# Can we get Survival for ALL the classes in one Query - Introducing GROUP BY

db.query('''SELECT pclass, count(*) as survival_count 
            from titanic 
            where survived = 1
            group by pclass;
        ''')
```

```python
# Can we also get how many did not survive

db.query('''SELECT pclass, survived, count(*) as survival_count 
            from titanic 
            group by pclass, survived;
        ''')
```

# Challenge Query

Get Artists that have more than 5 Records

```python
db_demo.query('''select artist.ArtistId, artist.name, count(*) 
                from artist, album 
                where artist.artistid = album.artistid
                group by artist.artistid, artist.name
                having count(*) > 5;
        ''')
```
